4 a)

By Jensen's inequality, $\ell_{semi-sup}(\theta^{(t+1)})$$\geq \sum_{i=1}^m ELBO(x^{(i)}; Q_i^{(t)}; \theta^{(t+1)})+\alpha\ell_{sup}(\theta^{(t+1)})$$\geq\sum_{i=1}^m ELBO(x^{(i)};Q_i^{(t)};\theta^{(t)})+\alpha\ell_{sup}(\theta^{(t)})$$=\ell_{unsup}(\theta^{(t)})+\alpha\ell_{\sup}(\theta^{(t)})$$=\ell_{semi-sup}(\theta^{(t)})$.

4 b)

In the E-step for a semi-supervised GMM, re-estimate latent variables for the first $m$ unlabelled samples by setting $Q_i^{(t)}(z^{(i)})$$=p(z^{(i)}|x^{(i)};\theta^{(t)})$$=\frac{p(z^{(i)};x^{(i)};\theta^{(t)})}{\sum_z p(z;x^{(i)};\theta^{(t)})}$$=\frac{p(x^{(i)}|\theta^{(t)};z^{(i)})p(z^{(i)})}{\sum_zp(x^{(i)}|\theta^{(t)};z)p(z)}$$=\frac{\mathcal{N}(\mu_{z^{(i)}},\Sigma_{z^{(i)}})(x^{(i)})\phi_{z^{(i)}}}{\sum_z(\mathcal{N}(\mu_{z^{(i)}},\Sigma_{z})(x^{(i)})\phi_{z})}$ for $i=1,...,m$.

4 c)

In the M-step, re-estimate $\phi_j$, $\mu_j$, and $\Sigma_j$ for $j=1,...,k$. Let $w_j^{(i)}=Q_i(z_j)$. The optimization objective equals $\ell_{semi-sup}(\theta)=\sum_{i=1}^m\left(\sum_{j}w^{(i)}_j\log\frac{p(x^{(i)},z^{(i)};\theta)}{w^{(i)}_j}\right)+\alpha\left(\sum_{i=1}^{\tilde{m}}\log p(\tilde{x}^{(i)},\tilde{z}^{(i)};\theta)\right)$$=\sum_{i=1}^m(\sum_{j=1}^kw_j^{(i)}\log\frac{\frac{1}{(2\pi)^d|\Sigma_j|^{1/2}}exp(-\frac{1}{2}(x^{(i)}-\mu_j)^T\Sigma_j^{-1}(x^{(i)}-\mu_j)\cdot\phi_j}{w_j^{(i)}})$$+\alpha(\sum_{i=1}^{\tilde{m}}\log (\frac{1}{(2\pi)^d|\Sigma_{j_{(i)}}|^{1/2}}\exp(-\frac{1}{2}(\tilde{x}^{(i)}-\mu_{j_{(i)}})\Sigma_{j_{(i)}}^{-1}(\tilde{x}^{(i)}-\mu_{j_{(i)}}))\cdot\phi_{j_{(i)}})).$

To satisfy $\sum_j\phi_j=1$, construct the Lagrangian $\mathcal{L}(\phi)=\left(\sum_{i=1}^m\sum_jw_j^{(i)}\log \phi_j+\alpha(\sum_{i=1}^{\tilde{m}}\log \phi_{j_{(i)}})\right)+\beta(\sum_{j=1}^k\phi_j-1).$ $\frac{\partial\mathcal{L}}{\partial\phi_j}=\sum_{i=1}^m \frac{w_j^{(i)}}{\phi_j}+\alpha(\sum_{i=1}^{\tilde{m}}\frac{1(j(i)=j)}{\phi_j})+\beta.$ $\frac{\partial\mathcal{L}}{\partial\phi_j}=0$ implies $\phi_j=\frac{\sum_{i=1}^mw_j^{(i)}+\alpha\sum_{i=1}^{\tilde{m}}1(j(i)=j)}{-\beta}.$ Since $\sum_{j=1}^k\phi_j=1$, $\beta=-m-\alpha\tilde{m}$. So, $\phi_j=\frac{\sum_{i=1}^mw_j^{(i)}+\alpha\sum_{i=1}^{\tilde{m}}1(j(i)=j)}{m+\alpha\tilde{m}}.$

$\nabla_{\mu_l}\ell_{unsup}(\theta)=\sum_{i=1}^m w_l^{(i)}(\Sigma_l^{-1}x^{(i)}-\Sigma_l^{-1}\mu_l)$ by the differentiation calculation on page 159 of notes. $\nabla_{\mu_l}\ell_{sup}(\theta)=\frac{\partial}{\partial\mu_l}(\sum_{i=1}^{\tilde{m}}-\frac{1}{2}(\tilde{x}^{(i)}-\mu_{j_{(i)}})\Sigma_{j_{(i)}}^{-1}(\tilde{x}^{(i)}-\mu_{j_{(i)}}))$$=\sum_{i=1}^{\tilde{m}}1(j_{(i)}=l)\frac{\partial}{\partial\mu_l}(-\frac{1}{2}(\tilde{x}^{(i)}-\mu_{l})\Sigma^{-1}_{l}(\tilde{x}^{(i)}-\mu_{l}))$$=\sum_{i=1}^{\tilde{m}}1(j_{(i)}=l)(\Sigma_{l}^{-1}\tilde{x}^{(i)}-\Sigma_l^{-1}\mu_l).$ So, $\nabla_{\mu_l}\ell_{semi-sup}(\theta)=\nabla_{\mu_l}\ell_{unsup}(\theta)+\alpha\nabla_{\mu_l}\ell_{sup}(\theta)$$=\sum_{i=1}^mw_l^{(i)}(\Sigma_l^{-1}x^{(i)}-\Sigma_l^{-1}\mu_l)+\alpha(\sum_{i=1}^{\tilde{m}}1(j_{(i)}=l)(\Sigma_l^{-1}\tilde{x}^{(i)}-\Sigma_l^{-1}\mu_l)).$ $\nabla_{\mu_l}\ell_{semi-sup}(\theta)=0$ implies $\sum_{i=1}^m w_l^{(i)}(x^{(i)}-\mu_l)+\alpha\sum_{i=1}^{\tilde{m}}1(j_{(i)}=l)(\tilde{x}^{(i)}-\mu_l)=0$. So, $\mu_l=\frac{\sum_{i=1}^m w_l^{(i)}x^{(i)}+\alpha\sum_{i=1}^{\tilde{m}}1(j_{(i)}=l)\tilde{x}^{(i)}}{\sum_{i=1}^m w_l^{(i)}+\alpha \sum_{i=1}^{\tilde{m}}1(j_{(i)}=l)}.$

$\nabla_{\Sigma_l}\ell_{unsup}(\theta)=\sum_{i=1}^m w_l^{(i)}\nabla_{\Sigma_l}(\log N(\mu_l,\Sigma_l)(x^{(i)}))$$=\sum_{i=1}^m w_l^{(i)}(-\frac{1}{2}(\Sigma_l^{-1}-\Sigma^{-1}_l(x^{(i)}-\mu_l)(x^{(i)}-\mu_l)^T\Sigma_l^{-1}))$. $\nabla_{\Sigma_l}\ell_{sup}(\theta)$$=\sum_{i=1}^{\tilde{m}}1(j_{(i)}=l)(-\frac{1}{2}(\Sigma_l^{-1}-\Sigma^{-1}_l(\tilde{x}^{(i)}-\mu_l)(\tilde{x}^{(i)}-\mu_l)^T\Sigma_l^{-1}))$. $\nabla_{\Sigma_l}\ell_{semi-sup}(\theta)=\nabla_{\Sigma_l}\ell_{unsup}(\theta)+\alpha\nabla_{\Sigma_l}\ell_{sup}(\theta)$$=\sum_{i=1}^m w_l^{(i)}(-\frac{1}{2}(\Sigma_l^{-1}-\Sigma^{-1}_l(x^{(i)}-\mu_l)(x^{(i)}-\mu_l)^T\Sigma_l^{-1}))$$+\alpha\sum_{i=1}^{\tilde{m}}1(j_{(i)}=l)(-\frac{1}{2}(\Sigma_l^{-1}-\Sigma^{-1}_l(\tilde{x}^{(i)}-\mu_l)(\tilde{x}^{(i)}-\mu_l)^T\Sigma_l^{-1})).$ $\nabla_{\Sigma_l}\ell_{semi-sup}(\theta)=0$ implies $\sum_{i=1}^m w_l^{(i)}(\Sigma_l-(x^{(i)}-\mu_l)(x^{(i)}-\mu_l)^T)+$$\alpha\sum_{i=1}^{\tilde{m}}1(j_{(i)}=l)(\Sigma_l-(\tilde{x}^{(i)}-\mu_l)(\tilde{x}^{(i)}-\mu_l)^T)$. This implies $\Sigma_l=\frac{\sum_{i=1}^m w_j^{(i)}(x^{(i)}-\mu_l)(x^{(i)}-\mu_l)^T+\alpha\sum_{i=1}^{\tilde{m}}1(j_{(i)}=l(\tilde{x}^{(i)}-\mu_l)(\tilde{x}^{(i)}-\mu_l)^T}{\sum_{i=1}^m w_l^{(i)}+\sum_{i=1}^{\tilde{m}}1(j_{(i)}=l)}.$
