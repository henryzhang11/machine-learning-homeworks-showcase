Suppose there are $2$ fixed points $V_1$ and $V_2$, then $B(V_1)=V_1$ and $B(V_2)=V_2$. $||B(V_1)-B(V_2)||_\infty=||V_1-V_2||_\infty$. So, $||V_1-V_2||_\infty\leq \gamma||V_1-V_2||_\infty$. So, $||V_1-V_2||_\infty=0$. So, $V_1=V_2$.
