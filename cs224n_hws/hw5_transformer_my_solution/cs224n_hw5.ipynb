1 a)

i) Viewing values $\{v_1,...,v_n\}$ as different categories and $\alpha$ as probabilities assigned to different categories, $c$ can be interpreted as a categorical probability distribution.
ii) If $\alpha_j$ is significantly larger than the other elements of $\alpha$, $k_jq$ is significantly larger than other $k_iq$s. So, $q$ must be approximately parallel to $k_j$ and approximately vertical to other $k_i$s.
iii) If $\alpha_j$ is significantly larger than the other weights, $c$ would be approximately parallel to $v_j$.
iv) Intuitively, the $q$'s contextual information is mostly contained by $x_j$.

1 b)

i) To extrct $v_a$ from $v=v_a+v_b$, project $v=v_a+v_b$ to $Span\{a_1,...,a_m\}$ by taking dot product with $a_1,...,a_m$ and then multiplying $a_1,...,a_n$ with the corresponding weights. Let $A=(a_1,...,a_m)$, then $AA^Tv$ gives the projection. So, $M=AA^T$.

ii) A query vector $q$ s.t. $c\approx\frac{1}{2}(v_a+v_b)$ satisfies $q\cdot k_a=q\cdot k_b>>0$ and $q\cdot k_i\approx 0$ for other $i$s. Since all $k_j$s are orthonormal to each other, let $q=k(v_a+v_b)$ for some $k>>0$.

1 c)

i) Set $q=k(\mu_a+\mu_b)$ for $k>>0$. $q k_i\sim N(k, \sum_{i=1}^d q_i^2 \alpha)$ for $i=a,b$ and $q k_i\sim N(0,\sum_{i=1}^d q_i\alpha)$ for other $i$s.

ii) Set $q=k(\mu_a+\mu_b)$, then $qk_i$ doesn't change for $i\neq a$. However, $qk_a=N(k, \sum_{i=1}^d q_i^2\alpha + \frac{1}{2}k^2)$. When $qk_a$ is very large, $c$ approximately equals $v_a$. When $qk_a$ is very small, $c$ approximately equals $v_b$.

1 d)

i) Let $q_1=k\mu_a$ and $q_2=k\mu_b$ for $k>>0$. This way $c_1=v_a$ approximately and $c_2=v_b$ approximately. So, $c=\frac{1}{2}(v_a+v_b)$ approximately.

ii) $q_1k_i=N(0,\sum_{i=1}^d (q_1)_i^2 \alpha)$ for $i\neq a$ and $q_1k_a=N(k, \sum_{i=1}^d (q_1)_i^2\alpha+\frac{1}{2}k^2)$. $q_2k_i=N(0,\sum_{i=1}^d (q_2)_i^2\alpha)$ for $i\neq b$ and $q_2k_b=N(k, \sum_{i=1}^d (q_2)_i^2\alpha)$. So, $c_2=v_b$ and $c_1$ oscillates between $v_a$ and potentially other combinations of $v_i$. So, $c=\frac{1}{2}(c_1+c_2)$ oscillates between $\frac{1}{2}(v_a+v_b)$ and $\frac{1}{2}(v+v_b)$ for $v$ being some random average of various $v_i$s.
